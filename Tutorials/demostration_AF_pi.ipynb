Demonstration Antiferromagnetic state has a Struct factor as find here. 
expressing the spin spin observable as -1 for the neighboring sites and 1 for the next neighboring sites I can split up the sum in 2 different sums of even and odd exponents with opposite signs 

```python
import numpy as np
import matplotlib.pyplot as plt

N = 100
q_vals = np.linspace(0, 2*np.pi, 100)
S = np.zeros_like(q_vals, dtype=complex)

for idx, q in enumerate(q_vals):
    even_sum = sum(np.exp(1j * q * (2*n)) for n in range(N//2))
    odd_sum  = sum(np.exp(1j * q * (2*n+1)) for n in range(N//2))
    S[idx] = (1/N) * (even_sum - odd_sum)


plt.plot(q_vals, np.abs(S))
plt.xlabel('q')
plt.ylabel('|S(q)|')
plt.xticks([0, np.pi, 2*np.pi], ['0', 'π', '2π'])
```
